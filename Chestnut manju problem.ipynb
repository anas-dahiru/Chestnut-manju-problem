```python
import math
import matplotlib.pyplot as plt

CHESTNUT_DIAMETER_CM = 2.54
CHESTNUT_RADIUS_CM = CHESTNUT_DIAMETER_CM / 2
CHESTNUT_VOLUME_CM3 = (4/3) * math.pi * (CHESTNUT_RADIUS_CM ** 3)
CHESTNUT_VOLUME_M3 = CHESTNUT_VOLUME_CM3 * 1e-6 

SOLAR_SYSTEM_RADIUS_METERS = 4.545e9
SOLAR_SYSTEM_VOLUME_M3 = (4/3) * math.pi * (SOLAR_SYSTEM_RADIUS_METERS ** 3)

volumes = []
minutes = []

def calculate_doubling_times(initial_volume, target_volume):
    doubling_times = 0
    while initial_volume < target_volume:
        initial_volume *= 2
        volumes.append(initial_volume)
        doubling_times += 5
        minutes.append(doubling_times)
    return doubling_times
time_to_fill_solar_system = calculate_doubling_times(CHESTNUT_VOLUME_M3, SOLAR_SYSTEM_VOLUME_M3)
print(f"it will take (time_to_fill_solar_system) minutes to cover the solar system with chestnut buns.")
plt.figure(figsize=(10,6))
plt.plot(minutes, volumes)
plt.xlabel("time(minutes)")
plt.ylabel("volume (cubic meters)")
plt.title("chestnut bun volume increase over time")
plt.grid(True)
plt.show()
```
